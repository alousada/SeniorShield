### NEISS CPS CODING MANUAL ANALYSIS

```python
import docx
import pandas as pd

# Open the .docx file
doc = docx.Document('2023NEISSCPSCCodingManualfinal.docx')

# Initialize a list to store the dataframes of the tables
tables_dataframes = []

# Initialize a DataFrame to store the combined data from tables 26 to 42
Table_product = pd.DataFrame()

# Initialize a DataFrame to store the table 22 - location related situation of the fall
Table_location = pd.DataFrame()

# Initialize a DataFrame to store the table 9 - diagnosis
Table_diagnosis = pd.DataFrame()

# Initialize a DataFrame to store the table 9 - diagnosis
Table_bodypart = pd.DataFrame()

# Initialize a DataFrame to store the table 9 - diagnosis
Table_disposition = pd.DataFrame()

# Loop through the tables in the document
for i, table in enumerate(doc.tables):
    # Initialize lists to store the rows and columns of the table
    table_data = []
    for row in table.rows:
        # Extract data from each cell in the row
        table_row = [cell.text.strip() for cell in row.cells]
        table_data.append(table_row)

    # Convert the list of table data into a Pandas DataFrame
    df_codes = pd.DataFrame(table_data[1:], columns=table_data[0])

    # Add the DataFrame to the list of table DataFrames
    tables_dataframes.append(df_codes)

    # Check if this is the 'Table 9' DataFrame
    if i == 8:
        # Rename the DataFrame to 'Table_09'
        df_codes.columns = ['Diagnosis Code', 'Diagnosis Value']
        Table_diagnosis= pd.concat([Table_diagnosis, df_codes], axis=0, ignore_index=True)

    # Check if this is the 'Table 22' DataFrame
    if i == 21:
        # Rename the DataFrame to 'Table_09'
        df_codes.columns = ['Code', 'Location']
        Table_location= pd.concat([Table_location, df_codes], axis=0, ignore_index=True)

    # Check if this is a table in the range 26 to 42
    if 26 <= i <= 42:
        # Append the data to the 'table_product' DataFrame
        Table_product = pd.concat([Table_product, df_codes], axis=0, ignore_index=True)

    # Check if this is a table is 13 'Table 14', body part
    if i == 13:
        # Append the data to the 'table_bodypart' DataFrame
        Table_bodypart = pd.concat([Table_bodypart, df_codes], axis=0, ignore_index=True)

    # Check if this is a table is 13 'Table 14', body part
    if i == 15:
        # Append the data to the 'table_bodypart' DataFrame
        Table_disposition = pd.concat([Table_disposition, df_codes], axis=0, ignore_index=True)



```

```python
def attributes_profile(df):
    import pandas as pd
    import matplotlib.pyplot as plt
    
    column_names = [
        'sex',
        'race',
        'hispanic',
        'diagnosis',
        'diagnosis_2',
        'body_part',
        'body_part_2',
        'disposition',
        'location',
        'fire_involvement',
        'alcohol',
        'drug',
        'product_1',
        'product_2',
        'product_3'
    ]
    
    # Seting a dictionary
    stacked_data = {}
    
    # Normalizing the values on stacked_data
    for column_name in column_names:
        normalized_values = df[column_name].value_counts(normalize=True)
        sorted_values = normalized_values.sort_values(ascending=False)
        stacked_data[column_name] = sorted_values
    
    # Sorting by the maximum frequency on each column
    stacked_data_sorted = {key: value for key, value in sorted(stacked_data.items(), key=lambda item: max(item[1]), reverse=True)}
    
    # Stacked bar graphic
    stacked_df = pd.DataFrame(stacked_data_sorted)
    stacked_df.transpose().plot(kind='bar', stacked=True, figsize=(11, 6), legend=False)
    
    plt.title('Key Attributes Profile')
    plt.xlabel('Category')
    plt.ylabel('Percentage')
    
    plt.tight_layout()
    plt.show()
```

### Variables mapping

Functions for mapping values to be used as necessary

```python
def add_names(df, column, mapping):
    """
    Substitui os cdigos numricos na coluna especificada do DataFrame pelos nomes correspondentes usando o mapeamento fornecido.
    """
    df[column] = df[column].replace(mapping[column])
    return df

```

```python
def remove_names(df, column, mapping):
    """
    Substitui os nomes na coluna especificada do DataFrame de volta para os cdigos numricos correspondentes usando o mapeamento fornecido.
    """
    # Crie um mapeamento reverso
    reverse_mapping = {v: k for k, v in mapping[column].items()}

    # Substitua os nomes pelos cdigos correspondentes
    df[column] = df[column].replace(reverse_mapping)
    return df
```

```python
import matplotlib.pyplot as plt

def starting_plots(df, sex_filter, race_filter, json_mapping):
    tag_1 = True  # Varivel de controle para a coluna "sex"
    tag_2 = True  # Varivel de controle para a coluna "race"

    if sex_filter != "all":
        df = df[df['sex'] == sex_filter]
        tag_1 = False

    if race_filter != "all":
        df = df[df['race'] == race_filter]
        tag_2 = False

    # DataFrames to plot
    dataframes = [df]
    panel_titles = ['First data view']

    for panel_idx, dataframe in enumerate(dataframes):
        # Determine the number of columns to plot and the number of rows and columns for subplots
        num_columns = 12  # Adjust this number based on the columns you want to plot
        num_rows = (num_columns - 1) // 3 + 1
        num_cols = min(num_columns, 3)

        # Create a panel chart with multiple visualizations
        fig, axs = plt.subplots(num_rows, num_cols, figsize=(15, 10))

        # Flatten the axs array if it's multidimensional
        axs = axs.ravel()

        # Histogram for the "age" column
        axs[0].hist(dataframe['age'], bins=20, color='skyblue', edgecolor='black')
        axs[0].set_title(f'{panel_titles[panel_idx]} - Age Histogram')

        # Bar charts for other numeric columns
        columns = ['diagnosis', 'location', 'body_part', 'disposition', 'fire_involvement', 'alcohol', 'drug', 'product_1', 'product_2', 'product_3']

        if tag_1:
            columns.insert(0, 'race')  # Adicione a coluna "sex" no incio da lista se o filtro for aplicado

        if tag_2:
            columns.insert(0, 'sex')  # Adicione a coluna "race" no incio da lista se o filtro for aplicado

        for i, column in enumerate(columns):
            if i + 1 < num_rows * num_cols:
                # Create a bar chart with the bars colored based on the maximum value
                value_counts = dataframe[column].value_counts()
                max_value = value_counts.max()
                colors = ['red' if count == max_value else 'skyblue' for count in value_counts]
                
                bars = axs[i + 1].bar(value_counts.index, value_counts, color=colors, edgecolor='black')
                axs[i + 1].set_title(f'{panel_titles[panel_idx]} - {column} Bar Chart')
                
                # Anote o nome do valor na barra vermelha com altura menor
                for bar, count, label in zip(bars, value_counts, value_counts.index):
                    if count == max_value and json_mapping.get(column):
                        axs[i + 1].annotate(json_mapping[column].get(label, ""), (bar.get_x() + bar.get_width() / 2, bar.get_height() - 5), ha='center', va='top', fontsize=6)


        # Adjust the layout
        plt.tight_layout()

        # Show the panel chart
        plt.show()

# Exemplo de uso:
# starting_plots(df, 'male', 'white', json_mapping)
```

```python
import plotly.graph_objects as go

# Calculate the count of unique combinations of body_part, diagnostic, and location
combinations = df_DOA.groupby(['diagnosis', 'location']).size().reset_index(name='Count')

# Map diagnosis values to names using Table_diagnosis
diagnosis_name_mapping = Table_diagnosis.set_index('Diagnosis Value')['Diagnosis Code'].to_dict()
combinations['diagnosis_label'] = combinations['diagnosis'].map(diagnosis_name_mapping)

# Map location values to names using Table_location
location_name_mapping = Table_location.set_index('Code')['Location'].to_dict()
combinations['location_label'] = combinations['location'].map(location_name_mapping)

# Calculate the percentual contribution for each diagnosis axis relative to df_DOA
combinations['Percentual_Diagnosis'] = (combinations.groupby('diagnosis')['Count'].transform('sum') / df_DOA.shape[0]) * 100

# Calculate the percentual contribution for each location axis relative to df_DOA
combinations['Percentual_Location'] = (combinations.groupby('location')['Count'].transform('sum') / df_DOA.shape[0]) * 100

# Sort combinations by Count (from highest to lowest)
combinations = combinations.sort_values(by='Count', ascending=False)

# Create unique labels for nodes (diagnosis and location) with values, names, and percentual contributions
unique_diagnoses_labels = [
    f"{value} ({diagnosis_name_mapping.get(value)})\n{combinations[combinations['diagnosis'] == value]['Percentual_Diagnosis'].values[0]:.2f}%"
    for value in combinations['diagnosis']
]
unique_locations_labels = [
    f"{value} ({location_name_mapping.get(value)})\n{combinations[combinations['location'] == value]['Percentual_Location'].values[0]:.2f}%"
    for value in combinations['location']
]
node_labels = unique_diagnoses_labels + unique_locations_labels

# Create unique nodes
unique_nodes = list(set(node_labels))

# Create a Sankey diagram
fig = go.Figure(go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.5),
        label=unique_nodes  # Use unique node labels
    ),
    link=dict(
        source=[unique_nodes.index(label) for label in unique_diagnoses_labels] * len(unique_locations_labels) +
               [unique_nodes.index(label) for label in unique_locations_labels] * len(unique_diagnoses_labels),
        target=[unique_nodes.index(label) for label in unique_locations_labels] * len(unique_diagnoses_labels) +
               [unique_nodes.index(label) for label in unique_diagnoses_labels] * len(unique_locations_labels),
        value=combinations['Count']
    )
))

# Customize the layout and appearance, including width and height
fig.update_layout(
    title="Sankey Diagram for Disposition 8 (Death on Arrival - DOA)",
    margin=dict(l=50, r=50, b=50, t=100),
    width=900,  # Adjust the width here
    height=600  # Adjust the height here
)

# Add annotations to display percentual contributions on each link
font_size = 10  # Adjust the font size here
for i, value in enumerate(combinations['Percentual_Diagnosis']):
    fig.add_annotation(
        x=1.1,  # Adjust the X-coordinate to move the labels to the right
        y=i,
        text=f"Diagnosis: {value:.2f}%",  # Text of the label for diagnosis axis
        font=dict(size=font_size),  # Font size
        showarrow=False
    )

for i, value in enumerate(combinations['Percentual_Location']):
    fig.add_annotation(
        x=1.1,  # Adjust the X-coordinate to move the labels to the right
        y=i + len(unique_diagnoses_labels),  # Adjust the y-coordinate for the location axis
        text=f"Location: {value:.2f}%",  # Text of the label for location axis
        font=dict(size=font_size),  # Font size
        showarrow=False
    )

# Show the Sankey diagram
fig.show()

```

```python
# Importe as bibliotecas necessrias
import re
import nltk
import pandas as pd
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Inicialize os recursos do NLTK se ainda no estiverem baixados
nltk.download('punkt')
nltk.download('stopwords')

# Funo para processamento de texto
def text_preprocessing(text):
    lemmatizer = WordNetLemmatizer()
    stop_words = set(stopwords.words('english'))

    # Remova caracteres especiais e converta para minsculas
    text = re.sub(r'\W', ' ', text.lower())

    # Remova nmeros nicos e espaos extras
    text = re.sub(r'\s+[a-zA-Z]\s+', ' ', text)

    # Remova nmeros nicos do incio
    text = re.sub(r'\^[a-zA-Z]\s+', ' ', text)

    # Substitua mltiplos espaos por um nico espao
    text = re.sub(r'\s+', ' ', text, flags=re.I)

    # Remova o prefixo 'b'
    text = re.sub(r'^b\s+', '', text)

    # Tokenize o texto
    words = nltk.word_tokenize(text)

    # Remova stopwords e aplique lematizao
    words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]

    # Reconstrua o texto aps o pr-processamento
    cleaned_text = ' '.join(words)

    return cleaned_text

# Pr-processamento de texto para seus DataFrames
X = df.copy()  # Crie uma cpia do DataFrame original

X['narrative_PD'] = X['narrative'].apply(text_preprocessing)

# Visualize as primeiras linhas do DataFrame aps o pr-processamento
print(X[['narrative', 'narrative_PD']].head())
```

```python
# Calculate the Jaccard similarity
def jaccard_similarity(set1, set2):
    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))
    return intersection / union
```

```python
# Extract and preprocess the words from each set of topics
def preprocess_topic(topic):
    # Remove numerical values and punctuations, and convert to lowercase
    topic_words = [word.strip('0123456789!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~').lower() for word in topic.split()]
    return set(topic_words)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cluster import KMeans
import nltk
from nltk.stem import WordNetLemmatizer
import re
import warnings

# Leting off warming alerts
warnings.filterwarnings("ignore", category=FutureWarning, module="seaborn")

# Text Vectorization using CountVectorizer
count_vectorizer = CountVectorizer()
X_counts = count_vectorizer.fit_transform(X['narrative_PD_ex'])

# Define a function to calculate the frequency of "cardiac" and similar words
def calculate_cardiac_similarity(text):
    cardiac_words = ['cardiac', 'heart', 'cardio', 'coronary', 'cardiopulmonary']  # Examples of similar words
    text = text.lower()
    count = sum(1 for word in cardiac_words if re.search(r'\b' + word + r'\b', text))
    return count

# Feature Engineering: Create a new column with the frequency of cardiac words
X['cardiac_frequency'] = [calculate_cardiac_similarity(text) for text in X['narrative_PD_ex']]

# K-Means Model Training
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X[['cardiac_frequency']])

# Label Assignment
X['cardiac_trend'] = kmeans.labels_

# Mapping the labels to 'yes' and 'no'
X['cardiac_trend'] = X['cardiac_trend'].map({0: 'no', 1: 'yes'})
```

```python
import re

def extract_keywords_(topics):
    keywords = []
    for topic_tuple in topics:
        # Extrai as palavras entre aspas duplas em cada tupla
        keywords.extend(re.findall(r'"([^"]*)"', topic_tuple[1]))
    return keywords
```

```python
def calculate_intersection(text, keywords):
    text_words = set(text.split())
    return len(text_words.intersection(keywords))
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Suponha que seu DataFrame seja chamado 'X'
# Selecionando as colunas relevantes para anlise
data = X[['age', 'sex', 'race', 'fatality_MAX']]

# Codificando variveis categricas (sexo)
data = pd.get_dummies(data, columns=['sex', 'race'], drop_first=True)

# Normalizao dos dados
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Calcula o nmero total de etapas
total_steps = 4  # 4 etapas no total (normalizao, PCA, t-SNE, K-Means)
current_step = 0  # Etapa atual

# Funo para imprimir o progresso
def print_progress():
    global current_step
    current_step += 1
    percent_complete = (current_step / total_steps) * 100
    print(f"Progresso: {percent_complete:.2f}%")

# Normalizao dos dados
print("Etapa 1: Normalizando os dados...")
data_scaled = scaler.fit_transform(data)
print_progress()

# Aplicar PCA para reduo de dimensionalidade
print("Etapa 2: Aplicando PCA para reduo de dimensionalidade...")
pca = PCA(n_components=0.95)  # Manter 95% da varincia
data_pca = pca.fit_transform(data_scaled)
print_progress()

# Reduo de dimensionalidade usando t-SNE no resultado do PCA
print("Etapa 3: Aplicando t-SNE para reduo de dimensionalidade...")
# Reduo de dimensionalidade usando t-SNE
tsne = TSNE(n_components=2, perplexity=30, n_iter=500, learning_rate=200, early_exaggeration=6, random_state=42)
data_tsne = tsne.fit_transform(data_pca)

print_progress()

# Aplicando o algoritmo K-Means aos dados reduzidos
print("Etapa 4: Aplicando K-Means aos dados reduzidos...")
kmeans = KMeans(n_clusters=3, random_state=42)
data['cluster'] = kmeans.fit_predict(data_pca)
print_progress()

# Visualizando os resultados
plt.figure(figsize=(12, 8))

# Grfico de disperso de t-SNE colorido por cluster
plt.scatter(data_tsne[:, 0], data_tsne[:, 1], c=data['cluster'], cmap='viridis')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.title('Clusters de t-SNE por Sexo, Raa e Idade')

plt.show()
```
